# Figure 1: Conceptual Figure 

Here we have visualize how simple functional forms of $\theta(N_0)$ affects the probability of rescue $P_\text{res} = 1 - \exp(-\theta(N_0)N_0)$

We depict three different functions 
1. $\theta(N_0) = \theta_1$, i.e., it is independent of $N_0$ (black line) 
2. $\theta(N_0) = \displaystyle \frac{\theta_1}{1+\theta_2 N_0}$, a decreasing function of $N_0$ (blue line). This starts at the same place as the black line but saturates at a much lower value. 
3. $\theta(N_0) = \displaystyle \frac{a_1 N_0}{1+a_2 N_0}$, is an increasing function of $N_0$ (red line). The values chosen are such that it start below the black line but increases to a value higher than the black line.

# Analytical value of rescue probability (Not a potential figure) 

This is the code for the exact resuce probability computed from theory. I use it to then simulate rescue probability over a finite number of replicates. I am doing this until we have finalized the parameters to run the full simulations for.

```python
def hyp2f1_pq(p,q,x):
    return hyp2f1(q,p,1+p,x)

def pest(x,bw,dw,bm,dm,R,gw,gm,N0): 
    rw = bw-dw 
    rm = bm-dm 
    Jt = (bm+dm)/rm * hyp2f1_pq(-rm/rw, -bm/bw, x) - bm/bw * x/(1-rm/rw) * hyp2f1_pq(1-rm/rw, 1-bm/bw,x) 
    Jt = Jt/  ( 1 -x )**(bm/bw)
    return 2*(1+Jt)**(-1)

def Pres(bw,dw,bm,dm,R,gw,gm,N0,mu) : 
    rw = bw-dw
    def integrand_alt(x): 
        return mu*R*pest(x,bw,dw,bm,dm,R,gw,gm,N0)/(gw*(1-x))    
    
    if R == np.inf: 
        Pres_density = 1-np.exp(mu*N0*(bm-dm)/(bm*(bw-dw)))
    else: 
        n0 = N0/R 
        Pres_density = 1- np.exp( - quad(integrand_alt, 0, n0/(n0-rw/gw))[0] )
    return Pres_density 
```

# Figure 2: Bacterial 

Here I fit two models to simulated data intended to match bacterial simulations. The exact parameters still need to be finalized but here I describe the statistical methods and potential plots. 

The two models I am fitting are (i) the null model $P_\text{res} = 1-\exp(-\theta N_0)$ and (ii) the alternative model $P_\text{res} = 1-\exp(-\displaystyle \frac{\theta_1}{1 + \theta_2 N_0} N_0)$. Note that this is a case of a nested model, so we are testing for if $\theta_2$ is significantly different from $0$ to reject the null model. 

To do this we fit the data to a generalized linear model (GLM) with a specific link functions. More simply, our data is a list of initial population sizes $N_{0,i}$ , the observed number of succesful rescue $y_i$ over a specified number of replicates $N_R$. Then the likelihood of observing the data given specific values of $\theta$s is 

$$ \displaystyle \sum_{i} {N_R \choose y_i} P_\text{res}^{y_i} (1-P_\text{res})^{N_R-y_i} $$ 

Now the $P_\text{res}$ here depends on $N_{0,i}$ and the $\theta$s.. 

We choose $\theta$ for the null model that minimizes the negative log-likelihood (-$L_\text{null}$) and $(\theta_1, \theta_2)$ that minimizes the negative log-likelihood (-$L_\text{alt}$) for the alternative model. Then we decide whether to reject the null model by looking at the deviance $D = 2( L_\text{alt} - L_\text{null})$. 

$D$ is supposed to be distributed according to a $\chi^2$ distribution with one degree of freedom. Then we can compute the probability $P(D < \chi^2)$.

```python
def model1(x, theta):
    return 1 - np.exp(-theta[0] * x)
def model2(x, theta):
    return 1 - np.exp(-theta[0] * x/(1+theta[1]*x) )


def neg_log_lik(model, theta, x, y, N):
    p = model(x,theta) 
    p = np.clip(p, 1e-12, 1-1e-12)
    
    ll = y * np.log(p) + (N - y) * np.log(1 - p)
    return -np.sum(ll)


def MLE_fit(Ndata_range, PRdata_range, model, theta0, bounds, NoR): 
    x = np.array(Ndata_range)
    N = np.array([NoR for xi in x])
    y = np.array([PR*NoR for PR in PRdata_range])

    res = minimize(
        lambda t: neg_log_lik(model, t, x, y, N),
        x0=theta0,
        bounds=bounds
    )

    theta = res.x
    
    PRpred_range = np.array([model(N, theta) for N in Ndata_range])
    
    ss_res = np.sum((PRdata_range - PRpred_range)**2)
    ss_tot = np.sum((PRdata_range - np.mean(PRdata_range))**2)
    r2 = 1 - ss_res/ss_tot
    
    
    return theta, r2, neg_log_lik(model, theta, x, y, N)

def alternate_test(N0range, PRdata, t0_init_range, t1_init_range, NoR):
    
    r2_1old = 0
    theta0 = [0.001]
    bounds = [(-1e9,None)]
    best_theta, best_nll1, best_r2_1 = None, np.inf, 0
    for t0_init in t0_init_range:
        theta0[0] = t0_init
        theta, r2_1, nll1 = MLE_fit(N0range, PRdata, model1, theta0, bounds, NoR)
        if nll1 < best_nll1:
            best_nll1 = nll1
            best_theta = theta
            best_r2_1 = r2_1
    theta1, r2_1, nll1 = best_theta, best_r2_1, best_nll1
    #theta, r2_1, nll1 = MLE_fit(N0range, PRdata, model1, theta0, bounds, NoR)
    #N_range = [ 10**x for x in np.arange(2,8,0.02) ] 
    #PR_range = [model1(N, theta1) for N in N_range]
    #ax[ind].plot(N_range, PR_range, color="C"+str(ind_R), alpha=0.5)

    r2_2old = 0
    theta0 = [0.001, 0]
    bounds = [(-1e9,None),(-1e-9,None)]
    best_theta2, best_nll2, best_r2_2 = None, np.inf, 0
    for t0_init in t0_init_range:
        theta0[0] = t0_init
        for t1_init in t1_init_range:
            theta0[1] = t1_init
            theta2, r2_2, nll2 = MLE_fit(N0range, PRdata, model2, theta0, bounds, NoR)
            if nll2 < best_nll2:
                best_nll2 = nll2
                best_theta2 = theta2
                best_r2_2 = r2_2
    theta2, r2_2, nll2 = best_theta2, best_r2_2, best_nll2

    #N_range = [ 10**x for x in np.arange(2,8,0.02) ] 
    #PR_range = [model2(N, theta1) for N in N_range]
    #ax[ind].plot(N_range, PR_range, color="C"+str(ind_R), linestyle = '--', alpha=0.5)

    LR = 2 * (-nll2 + nll1)
    p_value = chi2.sf(LR, df=1)
    
    return theta1, theta2, p_value, r2_1, r2_2, nll1, nll2

def saturated_test(N0range, PRdata, t0_init_range, NoR):
    theta0 = [0.001]
    bounds = [(-1e9, None)]
    
    best_theta, best_nll1, best_r2_1 = None, np.inf, 0
    
    for t0_init in t0_init_range:
        theta0[0] = t0_init
        theta, r2_1, nll1 = MLE_fit(N0range, PRdata, model1, theta0, bounds, NoR)
        
        if nll1 < best_nll1:
            best_nll1 = nll1
            best_theta = theta
            best_r2_1 = r2_1

    theta1, nll1, r2_1 = best_theta, best_nll1, best_r2_1

    x = np.array(N0range)
    N = np.array([NoR for _ in x])
    y = np.array([PR * NoR for PR in PRdata])

    p_sat = y / N
    p_sat = np.clip(p_sat, 1e-12, 1 - 1e-12)

    ll_sat = y * np.log(p_sat) + (N - y) * np.log(1 - p_sat)
    nll_sat = -np.sum(ll_sat)

    LR = 2 * ( -nll1 + nll_sat )

    df = len(N0range) - 1 
    p_value = chi2.sf(LR, df=df)

    return theta1, p_value, r2_1, nll1, nll_sat
    
```

Given a set of parameters we can run multiple replicates and get a distribution of $p$-values (or $R^2$ values) to determine if that is sufficient for detect deviations. Basically we can look at the proportion of $p$-values that are below 0.05 as a measure 

```python
def power_analysis_alternate(bw,dw,bm,dm,Rrange,gw,gm,N0range,mu,NoR,NoT): 
    plist = []
    for trial in range(NoT): 
        PRdata = [ np.random.binomial(NoR,Pres(bw,dw,bm,dm,Rrange[ind],gw,gm,N0range[ind],mu))/NoR for ind in range(len(N0range)) ]
        
        t0_init_range = [10**k for k in range(-9,-1,2)]
        t1_init_range = np.arange(0, 1.01, 0.2)
        theta1, theta2, p_value, r2_1, r2_2, nll1, nll2 = alternate_test(N0range, PRdata, t0_init_range, t1_init_range, NoR)
        plist += [p_value]
    
    return plist

def power_analysis_saturated(bw,dw,bm,dm,Rrange,gw,gm,N0range,mu,NoR,NoT): 
    plist =fig.suptitle("Power analysis", fontsize=15) []
    for trial in range(NoT): 
        PRdata = [ np.random.binomial(NoR,Pres(bw,dw,bm,dm,Rrange[ind],gw,gm,N0range[ind],mu))/NoR for ind in range(len(N0range)) ]
        
        t0_init_range = [10**k for k in range(-9,-1,2)]
        t1_init_range = np.arange(0, 1.01, 0.2)
        theta1, p_value, r2_1, nll1, nll_sat = saturated_test(N0range, PRdata, t0_init_range, NoR)
        plist += [p_value]
    
    return plist

```

## Figure 2b: Negative exponential form preserved

Here we explore the case where even with competition the negative exponential form is preserved. To do this, experimentally one would vary the the amount of resources and population size proportionally. One way to do this is the increase the volume of culture medium along with the inoculum size. 

Increasing the volume requires increasing the well size, so 96-well plates are not sufficient. We need 48 and 24 well plates. Further the handling time doesn't scale linearly. We don't go into too much detail here but use the following scheme which is equivalent to 5 96-well plates (480 wells). 3 96-well, 5 48-well and 8 24-well plates. This has a total of 720 (288+240+192) wells. We choose 8 population sizes with 80 replicates each (the remaining 80 wells are empty to keep the number of replicates constant for each population size but we can have different number of replicates for each pop size since the likelihood allows for that). 




```python
Vrange = [150, 175, 200, 300, 350, 400, 500, 1000]
N200 = 10**6 #Population size at a volume of 200
N0range = np.array(Vrange)*N200/200 
plt.figure(figsize=(4.7, 3.5))
labels = [r"No competition ($R = \infty$)", r"Weak competition ($ \frac{N_0}{R} = 1$)",r"Strong competition ($ \frac{N_0}{R} = 10$)"]
for ind_R, R200 in enumerate([np.inf, 10**6, 10**5]):
    NoR = 80
    Rrange = np.array(Vrange)*R200/200
    PRdata = [np.random.binomial(NoR,Pres(bw,dw,bm,dm,Rrange[ind],gw,gm,N0range[ind],mu))/NoR for ind in range(len(N0range)) ]
    
    theta1, p_value, r2_1, nll1, nll_sat = saturated_test(N0range, PRdata, t0_init_range, NoR)
    theta1_alt, theta2_alt, p_value_alt, r2_1_alt, r2_2_alt, nll1_alt, nll2_alt = alternate_test(N0range, PRdata, t0_init_range, t1_init_range, NoR)
    
    t0_init_range = [10**k for k in range(-9,-1,2)]
    t1_init_range = np.arange(0, 1.01, 0.2)
            
    N_range = [ 10**x for x in np.arange(2,8,0.02) ] 
    PR_range = [model1(N, theta1) for N in N_range]
    plt.plot(N_range, PR_range, color="C"+str(ind_R), alpha=0.6, linewidth=2 )
    
    N_range = [ 10**x for x in np.arange(2,8,0.02) ] 
    PR_range = [model2(N, theta2_alt) for N in N_range]
    plt.plot(N_range, PR_range, color="C"+str(ind_R), alpha=0.6, linewidth=2, linestyle = '--' )
    
    plt.scatter(N0range, PRdata, facecolors = 'none', edgecolors = "C"+str(ind_R), s=10, label = labels[ind_R])
    print(labels[ind_R] + " p-value = " + str(p_value) + ", R2 = " + str(round(r2_1,3) ))
    print(labels[ind_R] + " Alternate mode p-value = " + str(p_value_alt) + ", R2 = " + str(round(r2_2_alt,3) ))

plt.xscale('log')
plt.legend(loc="upper left", fontsize=9)
plt.xlabel(r"Initial Population Size $N_0$")
plt.ylabel(r"Rescue probability $P_\text{res}$")
plt.tight_layout()
plt.savefig("NEwCompetition.png", format='png', bbox_inches="tight")
```

```python
def power_analysis_alternate(bw,dw,bm,dm,Rrange,gw,gm,N0range,mu,NoR,NoT): 
    plist = []
    for trial in range(NoT): 
        PRdata = [ np.random.binomial(NoR,Pres(bw,dw,bm,dm,Rrange[ind],gw,gm,N0range[ind],mu))/NoR for ind in range(len(N0range)) ]
        
        t0_init_range = [10**k for k in range(-9,-1,3)]
        t1_init_range = np.arange(0, 1.01, 0.3)
        theta1, theta2, p_value, r2_1, r2_2, nll1, nll2 = alternate_test(N0range, PRdata, t0_init_range, t1_init_range, NoR)
        plist += [p_value]
    
    return plist

def power_analysis_saturated(bw,dw,bm,dm,Rrange,gw,gm,N0range,mu,NoR,NoT): 
    plist = []
    for trial in range(NoT): 
        PRdata = [ np.random.binomial(NoR,Pres(bw,dw,bm,dm,Rrange[ind],gw,gm,N0range[ind],mu))/NoR for ind in range(len(N0range)) ]
        
        t0_init_range = [10**k for k in range(-9,-1,3)]
        t1_init_range = np.arange(0, 1.01, 0.3)
        theta1, p_value, r2_1, nll1, nll_sat = saturated_test(N0range, PRdata, t0_init_range, NoR)
        plist += [p_value]
    
    return plist

bw = 0.5 - 0.01 #np.log(0.95)/2
bm = 0.5 + 0.01 #np.log(1 + sm)/2
dw = 0.5 + 0.01 #np.log(0.95)/2
dm = 0.5 - 0.01 #np.log(1+ sm)/2
mu = 10**(-6)
gw = 0.01
gm = 0.01
N0range = [1e2,4e2,1e3,4e3,1e4,4e4,1e5,4e5]
labels = ["Weak competition", "Strong competition"]
plt.figure(figsize=(4.5,4))
for ind_R, R in enumerate([10**6,10**5]):
    print(R)
    Rrange = [R for N0 in N0range ]
    NoT=100
    detection_probability = []
    NoRrange = np.array([1,5,10,15])*40
    for NoR in NoRrange :
        print(NoR)
        plist = power_analysis_alternate(bw,dw,bm,dm,Rrange,gw,gm,N0range,mu,NoR,NoT)
        plist = np.array(plist)
        #print(plist)
        detection_probability += [ len(plist[plist < 0.05])/NoT ]
    plt.plot(NoRrange, detection_probability, label = labels[ind_R] )
plt.legend()
plt.xlabel("Number of replates per population size")
plt.ylabel("Detection probability")
```

So, for the same range of population sizes, how many replicates do we need to have a good detection probability. As can be see, when competition is weak, even with 600 replicates per population size, the detection probability is less than 30\%.
